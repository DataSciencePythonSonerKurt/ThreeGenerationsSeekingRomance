```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",100)
from datetime import datetime
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("profiles.csv")
```

```python
df=df.drop(columns=['essay0','essay1','essay2','essay3','essay4','essay5','essay6','essay7','essay8','essay9'],axis=1)
```

```python
df.describe().T
```

```python
df.describe(include=['O'])#sayısal veri olmayan sütunların istatistiği
```

```python
df['age'] = df.age.fillna(df['age'].median())
df['body_type'] = df.body_type.fillna('average')
df['diet'] = df.diet.fillna('no response given')
df['drinks'] = df.drinks.fillna('no response given')
df['drugs'] = df.drugs.fillna('no response given')
df['education'] = df.education.fillna('no response given')
df['ethnicity'] = df.ethnicity.fillna('no response given')
df['job'] = df.job.fillna('no response given')
df['location'] = df.location.fillna('no response given')
df['offspring'] = df.offspring.fillna('no response given')
df['orientation'] = df.orientation.fillna('no response given')
df['pets'] = df.pets.fillna('no response given')
df['religion'] = df.religion.fillna('no response given')
df['sex'] = df.sex.fillna('no response given')
df['sign'] = df.sign.fillna('no response given')
df['smokes'] = df.smokes.fillna('no response given')
df['speaks'] = df.speaks.fillna('no response given')
df['status'] = df.status.fillna('no response given')
```

```python
df['height'] = df.height.fillna(df['height'].mean())
df['income'] = df.income.fillna(df['income'].mean())
```

```python
df['last_online']=df['last_online'].str.split('-').str[0]
```

```python
df['last_online']=df['last_online'].fillna(df['last_online'].median())
```

```python
df['last_online']=df['last_online'].astype(int)
```

```python
df['last_online'].unique()
```

```python
df['offspring'] = df['offspring'].str.replace('doesn&rsquo;t', "doesn't")
df['sign'] = df['sign'].str.replace('doesn&rsquo;t', "doesn't")
```

```python
df['education'].replace({'working':'student','graduated from':'graduated','dropped out':'dropped'},inplace=True)
```

```python
df['education']=df['education'].str.split().str[0]
```

```python
df['education'].replace({'college/university':'graduated','two-year':'graduated','space':'graduated','masters':'graduated','high':'graduated','ph.d':'graduated','law':'graduated'},inplace=True)
```

```python
df['body_type'].unique()
```

```python
df['body_type']=df['body_type'].map({'fit':'sporty',
                                     'athletic':'sporty',
                                     'thin':'sporty',
                                     'skinny':'sporty',
                                    'average':'average',
                                    'curvy':'curvy', 
                                     'a little extra':'curvy',
                                     'full figured':'curvy',
                                    'overweight':'bad',
                                     'jacked':'bad',
                                     'used up':'bad',
                                     'rather not say':'bad'})
```

```python
df['speaks']=df['speaks'].astype('category')
```

```python
df=df.drop(columns=['location','offspring','ethnicity','sign','speaks','pets','religion','drinks','drugs','smokes'],axis=1)
df
```

```python
#yaş grubu sütunu ekliyoruz
df['age']=df['age'].astype(int)

epoch = []
for i in range(len(df)):
    if 17 < int(df["age"].iloc[i]) < 33:
        epoch.append(1)
    elif 32 < int(df["age"].iloc[i]) < 48:
        epoch.append(2)
    else:
        epoch.append(3)
df["epoch"] = epoch
```

```python
df['body_type']=df['body_type'].astype('category')
df['diet']=df['diet'].astype('category')
df['sex']=df['sex'].astype('category')
df['status']=df['status'].astype('category')
df['orientation']=df['orientation'].astype('category')
```
